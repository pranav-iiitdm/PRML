```python
import numpy as np
```

```python
r1 = int(input('Enter the no.of rows in matrix 1 : '))
c1 = int(input('Enter the no.of columns in matrix 1 : '))
print(f'Enter the {r1} x {c1} elements : ')

matrix1 = []
for i in range(0,r1):
    arr = []
    print(f'Enter the elements for row - {i + 1}')
    for j in range(0,c1):
        num = int(input())
        arr.append(num)
    matrix1.append(arr)

r2 = int(input('Enter the no.of rows in matrix 2 : '))
c2 = int(input('Enter the no.of columns in matrix 2 : '))
print(f'Enter the {r2} x {c2} elements : ')
    
matrix2 = []
for i in range(0,r2):
    arr = []
    print(f'Enter the elements for row - {i + 1}')
    for j in range(0,c2):
        num = int(input())
        arr.append(num)
    matrix2.append(arr)

mat1 = np.array(matrix1)
mat2 = np.array(matrix2)
```

```python
print('Enter : ')
print('1. Matrix Addition')
print('2. Matrix Subtraction')
print('3. Scalar Matrix Multiplication')
print('4. Elementwise Matrix Multiplication')
print('5. Matrix Multiplication')
print('6. Matrix Transpose')
print('7. Trace of a Matrix')
print('8. Solve System of Linear Equations')
print('9. Determinant')
print('10. Inverse')
print('11. Eigen Value and Eigen Vector')
print('12. Exit')

ch = 0

while(ch != 12):
    ch = int(input())
    if(ch == 1):
        if(r1 == r2 and c1 == c2):
            out = np.add(mat1, mat2)
            print(out)
        else:
            print('Incompatible Matrices (Sizes)')
    elif(ch == 2):
        if(r1 == r2 and c1 == c2):
            out = np.subtract(mat1, mat2)
            print(out)
        else:
            print('Incompatible Matrices (Sizes)')
    elif(ch == 3):
        scalar = int(input('Enter the scalar value to be multiplied : '))
        ch1 = int(input('Choose the matrix (1 or 2) : '))
        if(ch1 == 1):
            print(np.multiply(mat1,scalar))
        elif(ch == 2):
            print(np.multiply(mat2,scalar))
        else:
            print('ERROR !!')
    elif(ch == 4):
        out = np.multiply(mat1, mat2)
        print(out)
    elif(ch == 5):
        if(c1 == r2):
            out = np.dot(mat1, mat2)
            print(out)
        else:
            print('Incompatible Matrices')
    elif(ch == 6):
        ch1 = int(input('Choose the matrix (1 or 2) : '))
        if(ch1 == 1):
            print(np.transpose(mat1))
        elif(ch1 == 2):
            print(np.transpose(mat2))
        else:
            print('ERROR !!')
    elif(ch == 7):
        ch1 = int(input('Choose the matrix (1 or 2) : '))
        if(ch1 == 1):
            print(np.trace(mat1))
        elif(ch1 == 2):
            print(np.trace(mat2))
        else:
            print('ERROR !!')
    elif(ch == 8):
        ch1 = int(input('Choose the matrix (1 or 2) : '))
        if(ch1 == 1):
            n = int(input('Enter the no.of rows in matrix B'))
            m = int(input('Enter the no.of columns in matrix B'))
            if(n != r1 or m != 1):
                print('INVALID MATRIX !!')
            else:
                matrixB = []
                for i in range(0,n):
                    arr = []
                    print(f'Enter the elements for row - {i + 1}')
                    for j in range(0,m):
                        num = int(input())
                        arr.append(num)
                    matrixB.append(arr)
                matB = np.array(matrixB)
                print(np.linalg.solve(mat1,matB))
        elif(ch1 == 2):
            n = int(input('Enter the no.of rows in matrix B'))
            m = int(input('Enter the no.of columns in matrix B'))
            if(n != r2 or m != 1):
                print('INVALID MATRIX !!')
            else:
                matrixB = []
                for i in range(0,n):
                    arr = []
                    print(f'Enter the elements for row - {i + 1}')
                    for j in range(0,m):
                        num = int(input())
                        arr.append(num)
                    matrixB.append(arr)
                matB = np.array(matrixB)
                print(np.linalg.solve(mat2,matB))
        else:
            print('ERROR !!')
    elif(ch == 9):
        ch1 = int(input('Choose the matrix (1 or 2) : '))
        if(ch1 == 1):
            out = np.linalg.det(mat1)
            print(out)
        elif(ch1 == 2):
            out = np.linalg.det(mat2)
            print(out)
        else:
            print('ERROR !!')  
    elif(ch == 10):
        ch1 = int(input('Choose the matrix (1 or 2) : '))
        if(ch1 == 1 and np.linalg.det(mat1) != 0):
            out = np.linalg.inv(mat1)
            print(out)
        elif(ch1 == 2 and np.linalg.det(mat2) != 0):
            out = np.linalg.inv(mat2)
            print(out)
        else:
            print('ERROR !!')
    elif(ch == 11):
        ch1 = int(input('Choose the matrix (1 or 2) : '))
        if(ch1 == 1):
            eigenvalues, eigenvectors = np.linalg.eig(mat1)
            print(eigenvalues)
            print(eigenvectors)
        elif(ch1 == 2):
            eigenvalues, eigenvectors = np.linalg.eig(mat2)
            print(eigenvalues)
            print(eigenvectors)
        else:
            print('ERROR !!')
    elif(ch == 12):
        print('Successfully Exited !!')
    else:
        print('Invalid Input !!')
    

   
```
