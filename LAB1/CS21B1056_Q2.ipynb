```python
import pandas as pd
import csv
import random
```

```python
def randomnumber(N):
    minimum = pow(10, N-1)
    maximum = pow(10, N) - 1
    return random.randint(minimum, maximum)
```

```python
data = {"Name" : ["Ram", "Sam", "Prabhu"],
        "AccountNumber" : [9893893891, 9893893898, 9893893871],
        "AccountType" : ["SB", "CA", "SB"],
        "Aadhaar_No" : [959389389173, 959389389179, 959389389159],
        "Balance" : [8989839, 7690990, 989330]}

df = pd.DataFrame(data)

df.to_csv('SBIAccountHolder.csv', index = False)
```

```python
print("Enter : ")
print("1. To Append Record of Account Holder")
print("2. To Delete a record")
print("3. To Credit Amount")
print("4. To Debit Amount")
print("5. To get Account Details")
print("6. To Exit")

ch = 0

while(ch != 6):
    ch = int(input())
    if(ch == 1):
        name = input('Enter the Account Holder Name : ')
        acc_no = randomnumber(10)
        selected_row = df[df['AccountNumber'] == acc_no]
        while(not selected_row.empty):
            acc_no = randomnumber(10)
            selected_row = df[df['AccountNumber'] == acc_no]
        acc_type = input("Enter the Account Type (SB or CA) : ")
        if(acc_type == 'SB' or acc_type == 'CA'):
            aadh_no = int(input('Enter Aadhar Number : '))
            balance = 0
            new_data = {"Name" : name, "AccountNumber" : acc_no,
                    "AccountType" : acc_type, "Aadhaar_No" : aadh_no,
                    "Balance" : balance}
            df = df.append(new_data, ignore_index = True)

            df.to_csv('SBIAccountHolder.csv', index = False)
        else:
            print("Invalid Account type")
    elif(ch == 2):
        acc_no = int(input("Enter the Account Number to Delete : "))
        selected_row = df[df['AccountNumber'] == acc_no].index
        if(not selected_row.empty):
            df.drop(selected_row, inplace = True)
        else:
            print("Invalid Account Number")
        df.to_csv('SBIAccountHolder.csv', index = False)
    elif(ch == 3):
        acc_no = int(input("Enter the Account Number : "))
        amount = int(input("Enter the amount to be credited :"))
        account = df[df["AccountNumber"] == acc_no]
        if(len(account) == 0):
            print("Enter Valid Account Number")
        if(amount <= 0):
            print("INVALID amount")
        else:
            df.loc[df[df['AccountNumber'] == acc_no].index, ['Balance']] += amount

        df.to_csv('SBIAccountHolder.csv', index = False)
    elif(ch == 4):
        acc_no = int(input("Enter the Account Number : "))
        account = df[df["AccountNumber"] == acc_no]
        amount = int(input("Enter the amount to be debited : "))
        curr_bal = df.loc[df[df['AccountNumber'] == acc_no].index, ['Balance']].values
        if(len(account) == 0):
            print("Enter Valid Account Number")
        if(amount <= 0):
            print("INVALID amount")
        else:
            if(curr_bal < amount and account["AccountType"].values[0] == 'SB'):
                print("Insufficient Balance")
            else:
                df.loc[df[df['AccountNumber'] == acc_no].index, ['Balance']] -= amount
        df.to_csv('SBIAccountHolder.csv', index = False)
    elif(ch == 5):
        acc_no = int(input("Enter the Account Number to get the details : "))
        selected_row = df[df['AccountNumber'] == acc_no]
        print(df[df['AccountNumber'] == acc_no])
        if(not selected_row.empty):
            selected_row = df[df['AccountNumber'] == acc_no]
            print("Account Holder Name : ",selected_row["Name"].values)
            print("Account Number : ",selected_row["AccountNumber"].values)
            print("Account Type : ",selected_row["AccountType"].values)
            print("Aadhaar Number : ",selected_row["Aadhaar_No"].values)
            print("Balance : ",selected_row["Balance"].values)
        else:
            print("Invalid Number")
    elif(ch == 6):
        print("EXITED SUCCESSFULLY !!")
    else:
        print("Invalid Input")
```
